# plotting the log data

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style("whitegrid")
```

```python
df=pd.read_csv("D:/Programming/DATA - SCIENCE/oil_data/Chapter 2/Matplotlib and Well Log Plotting/Chapter2_Sonic_Log_MIP_4H_DataSet.csv")
```

```python
df
```

```python
df=df[df.Well_Name=="MIP_4H"]
```

### Let’s plot depth (DEPT) on the y-axis and gamma ray (GR_EDTC), neutron porosity (NPHI), bulk density (RHOZ), and photoelectric (PEFZ) on the x axis

```python
fig,axes=plt.subplots(ncols=4,nrows=1,figsize=(10,10),sharey=True,dpi=100)
fig.suptitle("Log data vistualization ")
# figure 
y,x1,x2,x3,x4= df["DEPT"].values,df["GR_EDTC"].values,df["NPHI"].values,df["RHOZ"].values,df["PEFZ"].values
# figure_1
axes[0].plot(x1,y,c="black",linewidth=.5)
axes[0].set_title("GR",color="black",fontsize=15)
axes[0].xaxis.tick_top()

# figure_2
axes[1].plot(x2,y,c="blue",linewidth=.5)
axes[1].set_title("porosity",color="black",fontsize=15)
axes[1].xaxis.tick_top()
#figure_3
axes[2].plot(x3,y,c="red",linewidth=.5)
axes[2].set_title("denisty",color="black",fontsize=15)
axes[2].xaxis.tick_top()
#figure_4
axes[3].plot(x4,y,c="green",linewidth=.5)
axes[3].set_title("resistivity",color="black",fontsize=15)
axes[3].xaxis.tick_top()
```
