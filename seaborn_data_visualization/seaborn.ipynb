# jointplot( x= variable-1 , y=variable-2 , data= df )
### used to show the relation between two variables and the distribution of each variables also

```python
plt.figure()
sns.jointplot(x="petal_length",y="petal_width",data=iris)
plt.show()
```

#### from the above , the two variables has a positive correration

```python
plt.figure()
sns.jointplot(x="petal_length",y="petal_width",data=iris,kind="hex")
sns.set_style("whitegrid")
plt.show()
```

```python
plt.figure()
sns.jointplot(x="petal_length",y="petal_width",data=iris,kind="kde",color="blue",shade=True)
plt.show()
```

# nested plot

```python
import pandas as pd 
import numpy as np
```

```python
X=np.arange(1,20,2)
Y= [i**2 for i in X]
```

```python
fig=plt.figure(figsize=(10,6))
fig1=fig.add_axes([.1,.1,.8,.8])
fig2=fig.add_axes([.2,.5,.3,.2])
fig1.plot(X,Y,c="red",marker="+")
fig1.set_xlabel(" x ",fontsize=20,color="green")
fig1.set_ylabel(" y ",fontsize=20,color="green",rotation=0)
fig2.plot(X,Y,c="black",marker="o")
fig2.set_ylabel(" y ",fontsize=15,color="green",rotation=0)
fig2.set_xlabel(" x ",fontsize=15,color="green")
plt.show()
```

```python
plt.figure()
sns.lineplot(x=X,y=Y,palette="red")
sns.lineplot(x=Y,y=X)
plt.show()
```

# subplot using seaborn

```python
fig=plt.figure(figsize=(10,5))
fig1=fig.add_axes([.1,.1,.8,.8])
fig2=fig.add_axes([.2,.5,.3,.2])
fig1.plot(X,Y,c="red")
fig2.plot(X,Y,c="black")
plt.show()
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2,figsize=(8,8))
fig.suptitle("Suptitle")
sns.lineplot(x=X,y=Y,ax=axes[0,0])
axes[0,0].set_title("fuck")
sns.lineplot(x=X,y=Y,ax=axes[0,1])
axes[0,1].set_title("fuck")
sns.lineplot(x=X,y=Y,ax=axes[1,0])
axes[1,0].set_title("fuck")
sns.lineplot(x=X,y=Y,ax=axes[1,1])
axes[1,1].set_title("fuck")
```
