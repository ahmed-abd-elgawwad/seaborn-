# <span style="color:red"> data visualization using matplotlip </span>

## importing data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib notebook
```

```python
df= sns.load_dataset("tips")
```

```python
df
```

```python
df.describe()
```

## <span style="color:blue"> scatter plot </span>

```python
plt.figure()
plt.title("bill & size [vs] tip",fontsize=20,c="green",fontweight="bold")
plt.scatter(df.tip,df.total_bill,c="red", s=10,label="total_bill")
plt.scatter(df.tip,df["size"],c="blue",s=15,marker="+",label="size")
plt.legend()
plt.xlabel("Tip",fontsize=15,c="green",fontweight="bold")
plt.show()
```

```python
plt.figure()
plt.plot(df.tip,df.total_bill,c="red")
plt.show()
```

### we note that:-
#### 1- the plot is the same as scatter but with connecting the points.
#### 2- to make a new plot and not plotting oon the same figure we use  ( plt.figure() & plt.show() ) for each plot.

### <span style="color:green"> what is i want to know the day that has the max somker clints </span>

```python
df.groupby("day")["smoker"].value_counts()
```

```python
plt.figure()
sns.countplot(x="smoker",hue="day",data=df,palette="RdBu")
plt.show()
```

```python
plt.figure()
sns.countplot(df.smoker,palette="RdBu")
plt.show()
```

### <span style="color:green"> what is the day the has the income and what meal exactly </span>

```python
plt.figure()
df.groupby("day").total_bill.mean().plot(kind="bar")
plt.xticks(rotation=0,fontsize=10)
plt.show()
```

## xticks -> the values of the x label
## yticks -> the vales of the y label

```python
plt.figure()
df.groupby(["day","time"]).total_bill.agg(["mean","std"]).plot(kind="bar")
plt.xticks(fontsize=5,rotation=30)
plt.show()
```

# <span style="color:blue"> subplots </span>

```python
from sklearn import datasets
data= datasets.load_iris()
```

```python
df2=pd.DataFrame(data.data, columns=data.feature_names)
```

```python
%matplotlib inline
f,ax= plt.subplots(nrows=4,ncols=1,figsize=(10,15))
f.suptitle("columns vs the index", fontsize=20,fontweight="bold",color="green")
for i,col in enumerate(df2.columns):
    ax[i].plot(df2.index,df2[col],c="red")
    ax[i].set_xlabel("Index")
    ax[i].set_ylabel(col)
#     ax[i].set_title(f"{col} vs index")
```

## or simply plotting them in sigle plot 

```python
%matplotlib notebook
plt.figure(figsize=(10,10))
plt.title("index vs columns", fontsize=20, color="blue", fontweight="bold")
for i in df2.columns:
    plt.plot(df2.index,df2[i],label=i)
plt.legend()
plt.xlabel("index",fontsize=15)
plt.ylabel("columns",fontsize=15)
plt.show()
```
